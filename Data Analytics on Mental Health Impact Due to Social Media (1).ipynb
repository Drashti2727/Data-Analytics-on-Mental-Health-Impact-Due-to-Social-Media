### üîç CHECKING FOR OUTLIER'S

```python
from scipy import stats
import numpy as np
import pandas as pd
# SELECT ONLY NUMERICAL COLUMN
numerical_columns = df_social_data.select_dtypes(include=[np.number]).columns
print(f"üîç Checking for outliers in the following columns: {list(numerical_columns)}\n")
# COMPUTE Z-SCORE
z_scores = np.abs(stats.zscore(df_social_data[numerical_columns]))
# DATA FRAME TO STORE OUTLIER'S DETAIL
outlier_rows = pd.DataFrame()
outlier_summary = {}  # DICTIONARY TO TRACK WHICH COLUMN HAVE OUTLIER'S
# ITRATE THROUGH NUMERICAL COLUMNS
for column in numerical_columns:
    #FIND OUTLIER'S INDICES WHERE Z-SCORE > 3
    outlier_indices = df_social_data.index[z_scores[column] > 3].tolist()
    # STORE THE OUTLIER'S STATUS IN THE SUMMARY DICTIONARY
    outlier_summary[column] = len(outlier_indices) > 0
    # PRINT OUTLIER'S DETAIL IF FOUND
    if outlier_indices:
        print(f"üîπ **Column: {column}** - Outliers detected in {len(outlier_indices)} rows.")
        outlier_rows = pd.concat([outlier_rows, df_social_data.loc[outlier_indices]], axis=0)

# REMOVE DUPLICATE ROW
outlier_rows = outlier_rows.drop_duplicates()
# DISPLAY TOTAL COUNT OF OUTLIER'S
print(f"\n‚úÖ **Total outliers detected: {outlier_rows.shape[0]}**\n")
#PRINT SUMMARY OF COLUMNS CHECKED AND WHETHER OUTLIER'S WERE FOUND
for col, has_outliers in outlier_summary.items():
    status = "‚úÖ No outliers found" if not has_outliers else "‚ùå Outliers detected"
    print(f"üìä Column: {col} -> {status}")
# DISPLAY FIRST FEW OUTLIER'S ROW IF IT EXIST
if not outlier_rows.empty:
    print("\nüîç Sample of detected outliers:")
    print(outlier_rows.head())
else:
    print("\nüéâ No outliers detected in the dataset!")
```
